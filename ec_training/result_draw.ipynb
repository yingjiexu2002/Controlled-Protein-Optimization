```python
import matplotlib.pyplot as plt
import numpy as np

# 设置全局字体为times new roman和字体大小
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 16

x_font_size = 20
y_font_size = 20
title_font_size = 20
legend_font_size = 16

color_light_blue = '#389cbc'
color_deep_blue = '#2878b5'
color_orange = '#D87C7C'
color_green = '#549F9A'
```

```python
# p=0.5
# data = [0.9310344827586207, 0.8305084745762712, 0.6017699115044248 ,0.14728682170542637]
data = [0.9571428571428572,0.9025974025974026,0.46357615894039733,0.42105263157894735]

# 创建x轴的标签
labels = ['2.x.x.x', '2.7.x.x', '2.7.7.x', '2.7.7.6']
# 创建一个颜色列表
# colors = ['#7E99F4', '#E29135']
colors = ['#2878b5', '#2878b5', '#2878b5', '#2878b5']

# 创建一个figure和axes
fig, ax = plt.subplots()

# 创建柱状图，为每个数据指定一个颜色
rects = ax.bar(labels, data, color=colors, width=0.5)

# 添加x轴和y轴的标签
ax.set_xlabel('Control tag')
ax.set_ylabel('Accuracy')

# 添加标题
ax.set_title('Accuracy of the generated data')

# 显示图形
plt.show()
```

```python
# 绘制一个柱状图
# 每个横坐标对应三种类型的数据
labels = ['2.x.x.x', '2.7.x.x', '2.7.7.x', '2.7.7.6']

# 这是每个横坐标对应的三种类型的数据
data_p025 = [1.0, 0.9787446861715429, 0.9598214285714286, 0.9215025906735751]
data_p050 = [0.9997190221972464, 0.9347034339229969, 0.5576008273009307, 0.5130102040816327]
data_p075 = [0.9742011140428027, 0.8775728414862336, 0.4144, 0.38904823989569753]

x = np.arange(len(labels))  # 横坐标的位置
width = 0.2  # 柱子的宽度

fig, ax = plt.subplots()

# 在每个横坐标上绘制三个柱子
rects1 = ax.bar(x - width, data_p025, width, label='p=0.25',color=color_deep_blue, hatch='/')
rects2 = ax.bar(x, data_p050, width, label='p=0.50',color=color_orange, hatch='\\')
rects3 = ax.bar(x + width, data_p075, width, label='p=0.75',color=color_green, hatch='x')

# 设置x轴的刻度和标签
ax.set_xticks(x)
ax.set_xticklabels(labels)

# 添加x轴和y轴的标签
ax.set_xlabel('Groups')
ax.set_ylabel('Scores')

# 添加标题
ax.set_title('Scores by group and type')

# 添加图例
ax.legend()

# 显示图形
plt.show()

```
